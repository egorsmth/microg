```python
import re

def parse_custom_string(input_string):
    """
    Parses a custom formatted string into a tuple of values.

    Parameters:
    input_string (str): The string to be parsed.

    Returns:
    tuple: A tuple containing the parsed integer or float values.
    """
    # Use regular expressions to find all numbers in the string
    numbers = re.findall(r'[-+]?\d+\.?\d*', input_string)
    
    # Convert the extracted strings to appropriate types
    parsed_values = []
    for num in numbers:
        if '.' in num:  # Check if the number is a float
            parsed_values.append(float(num))
        else:           # Otherwise, it's an integer
            parsed_values.append(int(num))
    
    # Determine the type of the first value and return the tuple
    if input_string[3] == 'i':
        return tuple(parsed_values)
    else:
        return tuple(parsed_values)
```

```python
class Value:
    def __init__(self, data, label='', op='', parents=()):
        self.data = data
        self.label = label
        self.op = op
        self.parents = parents
        self.grad = 0.0
        self._backward = lambda: None

    def backprop(self, init=1.0):
        self.grad += init
        l = self.parents
        for k, v in enumerate(l):
            if self.op == '+':
                v.backprop(self.grad)
            elif self.op == '*':
                a = list(map(lambda x: x.data, l[:k]))
                b = list(map(lambda x: x.data, l[k+1:]))
                v.backprop(self.grad * sum(a + b))
            elif self.op == 'tanh':
                v.backprop(self.grad * (1 - self.data**2))
            elif self.op == 'exp':
                v.backprop(self.grad * self.data)
            elif self.op.startswith("**"):
                (r, d) = parse_custom_string(self.op)
                if r == 1.0:
                    v.backprop(self.grad * (math.log(self.data)))
                else:
                    v.backprop(self.grad * (r*(d**(r-1))))
    
    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other, label=f'gen_{random.randrange(100)}')
        out = Value(self.data + other.data, parents=(self, other), label=f'({self.label}+{other.label})', op='+')
        
        def _backward():
            self.grad += out.grad
            other.grad += out.grad
        out._backward = _backward
        
        return out
    def __radd__(self, other):
        return self + other
        
    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other, label=f'gen_{random.randrange(100)}')
        out = Value(self.data * other.data, parents=(self, other), label=f'({self.label}*{other.label})', op='*')

        def _backward():
            self.grad += other.data * out.grad
            other.grad += self.data * out.grad
        out._backward = _backward

        return out
    
    def __rmul__(self, other):
        return self * other

    def __pow__(self, k):
        assert(isinstance(k, (int, float))) ## only accepting int and float for now
        if isinstance(k, int):
            t = 'i'
        else:
            t = 'f'

        out = Value(self.data**k, parents=(self, ), label=f'{self.label}**{k}', op=f'**{t}{k} {self.data}')

        def _backward():
            self.grad += (k * self.data**(k-1)) * out.grad
        out._backward = _backward

        return out

    def __truediv__(self, other):
        return self * (other**-1)

    def __neg__(self):
        return self * Value(-1, label=f'(-{self.label})')

    def __sub__(self, other):
        return self + (-other)
    
    def tanh(self):
        x = self.data
        n = np.tanh(x)
        out = Value(n, parents=(self, ), label=f'tanh({self.label})', op='tanh')

        def _backward():
            self.grad += (1 - n**2) * out.grad
        out._backward = _backward

        return out

    # def exp(self):
    #     n = math.exp(self.data)
    #     return Value(math.exp(self.data), parents=(self, ), label=f'exp({self.label})', op='exp')

    def backward(self):

        # topological order all of the children in the graph
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v.parents:
                    build_topo(child)
                topo.append(v)
        build_topo(self)

        # go one variable at a time and apply the chain rule to get its gradient
        self.grad = 1
        for v in reversed(topo):
            v._backward()
        
    def __repr__(self):
        return f'Value(data={self.data})'
```

```python
from graphviz import Digraph


def bfs(node, nodes):
    nodes.add(node)
    for i in node.parents:
        bfs(i, nodes)

def show_graph(v, labels=False):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) 

    nodes = set()
    bfs(v, nodes)

    for i, n in enumerate(nodes):
        
        dot.node(n.label, "{ %s | data: %.4f | grad: %.4f }" % (n.label if labels else '', n.data, n.grad), shape='record')
        op_name = f'{n.label}_op'
        if n.parents != ():
            dot.node(op_name, n.op)
            dot.edge(op_name, n.label)
            for p in n.parents:
                dot.edge(p.label, op_name)
    
    return dot
```

```python
class Neuron:
    def __init__(self, nin, ln, nn, nond=True):
        def gen():
            return random.uniform(-1, 1) if nond else 0.7
        self.w = [Value(gen(), label=f'w_ln{ln}_nn{nn}_p{i}') for i in range(nin)]
        self.b = Value(0.0, label=f'b_ln{ln}_nn{nn}')
        
    def __call__(self, x):
        res = sum((xn*wn for wn, xn in zip(self.w, x)), self.b)
        out = res.tanh() ; out.label = f'tanh_{random.randrange(1000)}'
        return out
        
    def parameters(self):
        return self.w + [self.b]

class Layer:
    def __init__(self, nin, nout, ln):
        self.neurons = [Neuron(nin, ln=ln, nn=i) for i in range(nout)]
    def __call__(self, x):
        out = [n(x) for n in self.neurons]
        return out[0] if len(out) == 1 else out
    def parameters(self):
        return [p for neur in self.neurons for p in neur.parameters()]

class MLP:
    def __init__(self, nin, nouts):
        sz = [nin] + nouts
        self.layers = [Layer(sz[i], sz[i+1], i) for i in range(len(nouts))]
    def __call__(self, x):
        for l in self.layers:
            x = l(x)
        return x
    def parameters(self):
        return [p for layer in self.layers for p in layer.parameters()]

```

```python
o = MLP(3, [4, 4, 1])
xs = [
    [Value(2.0, label='x_0_0'), Value(3.0, label='x_0_1'), Value(-1.0, label='x_0_2')],
    [Value(3.0, label='x_1_0'), Value(-1.0, label='x_1_1'), Value(0.5, label='x_1_2')],
    [Value(0.5, label='x_2_0'), Value(1.0, label='x_2_1'), Value(1.0, label='x_2_2')],
    [Value(1.0, label='x_3_0'), Value(1.0, label='x_3_1'), Value(-1.0, label='x_3_2')]
]

ys = [1.0, -1.0, -1.0, 1.0]
ypred = [o(x) for x in xs]
loss = sum((yout-ygt)**2 for ygt, yout in zip(ys, ypred))   ; loss.label = 'loss' 

for i in range(15):
    for p in o.parameters():
        p.grad = 0.0
    loss.backward()
    for p in o.parameters():
        p.data += -0.1 * p.grad
    
    ypred = [o(x) for x in xs]
    loss = sum((yout-ygt)**2 for ygt, yout in zip(ys, ypred))
    # print(f'{i}, {loss}')

for i in range(15):
    for p in o.parameters():
        p.grad = 0.0
    loss.backward()
    for p in o.parameters():
        p.data += -0.01 * p.grad
    
    ypred = [o(x) for x in xs]
    loss = sum((yout-ygt)**2 for ygt, yout in zip(ys, ypred))
    print(f'{15 + i}, {loss}')
print(loss)
print(ypred)
```
