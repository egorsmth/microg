```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
class Value:
    def __init__(self, data, label='', op='', parents=()):
        self.data = data
        self.label = label
        self.op = op
        self.parents = parents
        self.grad = 0.0

    def backprop(self, init=1.0):
        self.grad += init
        l = self.parents
        for k, v in enumerate(l):
            if self.op == '+':
                v.backprop(self.grad)
            elif self.op == '*':
                a = list(map(lambda x: x.data, l[:k]))
                b = list(map(lambda x: x.data, l[k+1:]))
                v.backprop(self.grad * sum(a + b))
            elif self.op == 'tanh':
                v.backprop(self.grad * (1 - self.data**2))
            elif self.op == 'exp':
                v.backprop(self.grad * self.data)
            elif self.op.startswith("**"):
                t = self.op[2:3]
                if t == 'f':
                    r = float(self.op[3:])
                else:
                    r = int(self.op[3:])
                if r == 1:
                    v.backprop(self.grad * (math.log(self.data)))
                else:
                    v.backprop(self.grad * (r*(self.data**(1/r))**(r-1)))
                
    def __add__(self, other):
        o = other if isinstance(other, Value) else Value(other, label=f'gen_{random.randrange(100)}')  
        return Value(self.data + o.data, parents=(self, o), label=f'({self.label}+{o.label})', op='+')

    def __mul__(self, other):
        o = other if isinstance(other, Value) else Value(other, label=f'gen_{random.randrange(100)}')
        return Value(self.data * o.data, parents=(self, o), label=f'({self.label}*{o.label})', op='*')
    
    def __rmul__(self, other):
        return self * other

    def __pow__(self, k):
        assert(isinstance(k, (int, float))) ## only accepting int and float for now
        if isinstance(k, int):
            t = 'i'
        else:
            t = 'f'
        return Value(self.data**k, parents=(self, ), label=f'{self.label}**{k}', op=f'**{t}{k}')

    def __truediv__(self, other):
        return self * (other**-1)

    def __neg__(self):
        return self * Value(-1, label=f'(-{self.label})')

    def __sub__(self, other):
        return self + (-other)
    
    def tanh(self):
        x = self.data
        n = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        return Value(n, parents=(self, ), label=f'tanh({self.label})', op='tanh')

    def exp(self):
        return Value(math.exp(self.data), parents=(self, ), label=f'exp({self.label})', op='exp')
        
    def __repr__(self):
        return "{} = {}".format(self.label, self.data)
```

```python
from graphviz import Digraph


def bfs(node, nodes):
    nodes.add(node)
    for i in node.parents:
        bfs(i, nodes)

def show_graph(v):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) 

    nodes = set()
    bfs(v, nodes)

    for i, n in enumerate(nodes):
        
        dot.node(n.label, "{ %s | data: %.4f | grad: %.4f }" % (n.label, n.data, n.grad), shape='record')
        op_name = f'{n.label}_op'
        if n.parents != ():
            dot.node(op_name, n.op)
            dot.edge(op_name, n.label)
            for p in n.parents:
                dot.edge(p.label, op_name)
    
    return dot
```
